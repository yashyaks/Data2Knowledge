```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("India-2001.csv")
df2 = pd.read_csv("India-2002.csv")
df3 = pd.read_csv("India-2003.csv")
df4 = pd.read_csv("India-2004.csv")
df5 = pd.read_csv("India-2005.csv")
df6 = pd.read_csv("India-2006.csv")
df7 = pd.read_csv("India-2007.csv")
df8 = pd.read_csv("India-2008.csv")
df9 = pd.read_csv("India-2009.csv")
df10 = pd.read_csv("India-2010.csv")
df11 = pd.read_csv("India-2011.csv")
```

```python
df1['2001'] = df1['M'] + df1['F']
df2['2002'] = df2['M'] + df2['F']
df3['2003'] = df3['M'] + df3['F']
df4['2004'] = df4['M'] + df4['F']
df5['2005'] = df5['M'] + df5['F']
df6['2006'] = df6['M'] + df6['F']
df7['2007'] = df7['M'] + df7['F']
df8['2008'] = df8['M'] + df8['F']
df9['2009'] = df9['M'] + df9['F']
df10['2010'] = df10['M'] + df10['F']
df11['2011'] = df11['M'] + df11['F']
```

```python
df1.head()
```

```python
df1.drop(['M', 'F'], axis=1, inplace=True)
df2.drop(['M', 'F'], axis=1, inplace=True)
df3.drop(['M', 'F'], axis=1, inplace=True)
df4.drop(['M', 'F'], axis=1, inplace=True)
df5.drop(['M', 'F'], axis=1, inplace=True)
df6.drop(['M', 'F'], axis=1, inplace=True)
df7.drop(['M', 'F'], axis=1, inplace=True)
df8.drop(['M', 'F'], axis=1, inplace=True)
df9.drop(['M', 'F'], axis=1, inplace=True)
df10.drop(['M', 'F'], axis=1, inplace=True)
df11.drop(['M', 'F'], axis=1, inplace=True)
```

```python
df2.head()
```

```python
merged_df = df1.merge(df2, on='Age').merge(df3, on='Age').merge(df4, on='Age').merge(df5, on='Age').merge(df6, on='Age').merge(df7, on='Age').merge(df8, on='Age').merge(df9, on='Age').merge(df10, on='Age').merge(df11, on='Age')
merged_df
```

```python
merged_df.plot(x='2002', kind='bar', stacked=True,title='Stacked Bar Graph by dataframe')
```

```python
merged=merged_df.T
merged
```

```python
header_row=merged.loc['Age']
df=pd.DataFrame(merged.values[1:],columns=header_row)
df.head()
```

```python
new_col = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011]
df.insert(loc = 0,
          column = 'Year',
          value = new_col)
df
```

```python
df.plot(x='Year', kind='bar', stacked=True,title='Stacked Bar Graph by dataframe',figsize=(30,15))
plt.show()
```
